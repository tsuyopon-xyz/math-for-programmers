```python
from vector_drawing import *
from math import sin, cos, tan, pi, sqrt, asin, atan2
```

```python
def calc_distance(vector):
    return sqrt(vector[0] ** 2 + vector[1] ** 2)

# 極座標（例: (長さ, 角度)） から 直交座標（例: (x座標, y座標)）への変換
def to_cartesian(polar_vector):
    length, angle = polar_vector[0], polar_vector[1]
    return (length * cos(angle), length * sin(angle))

# 直交座標から極座標への変換
def to_polar(cartesian_vector):
    x, y = cartesian_vector[0], cartesian_vector[1]
    length = calc_distance(cartesian_vector)
    angle = atan2(y, x)
    return (length, angle)

# 回転（問2.42）
def rotate(angle, vectors):
    polars = [to_polar(v) for v in vectors]
    return [to_cartesian((l, a+angle)) for l,a in polars]

# 平行移動
def translate(translation, vectors):
    return [add(translation, v) for v in vectors]

def add(*vectors):
    return (sum(v[0] for v in vectors), sum(v[1] for v in vectors))
```

```python
dino_vectors = [
  (6,4), (3,1), (1,2),
  (-1,5), (-2,5), (-3,4),
  (-4,4), (-5,3), (-5,2),
  (-2,2), (-5,1), (-4,0),
  (-2,1), (-1,0), (0,-3),
  (-1,-4), (1,-4), (2,-3),
  (1,-2), (3,-1), (5,1)
]
```

```python
rotation_angle = pi / 4
rotated_dino_vectors = rotate(rotation_angle, dino_vectors)
translate_and_rotated_dino_vectors = translate((8, 8), rotate(pi * 2, dino_vectors))

draw(
    Polygon(*dino_vectors, color=gray),
    Polygon(*rotated_dino_vectors, color=red),
    Polygon(*translate_and_rotated_dino_vectors, color=blue),
)
```
