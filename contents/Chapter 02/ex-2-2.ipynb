```python
# 問2.6, 2,7
def add(*vectors):
    return (sum(v[0] for v in vectors), sum(v[1] for v in vectors))

# 問2.8
def translate(translation, vectors):
    return [add(translation, v) for v in vectors]

```

```python
u = (-2, 0)
v = (1.5, 1.5)
w = (4, 1)

uv = add(u, v)
vw = add(v, w)
uw = add(u, w)
uvw = add(u, v, w)

print(uv, vw, uw, uvw)
```

```python
dino_vectors = [
  (6,4), (3,1), (1,2),
  (-1,5), (-2,5), (-3,4),
  (-4,4), (-5,3), (-5,2),
  (-2,2), (-5,1), (-4,0),
  (-2,1), (-1,0), (0,-3),
  (-1,-4), (1,-4), (2,-3),
  (1,-2), (3,-1), (5,1)
]

#問2.11
def hundred_dinos():
    translations = [(12*x, 10*y)
                    for x in range(-5, 5)
                    for y in range(-5, 5)]
    dinos = [Polygon(*translate(t, dino_vectors), color=blue) for t in translations]
    draw(*dinos, grid=None, axes=None, origin=None)
```

```python
hundred_dinos()
```
