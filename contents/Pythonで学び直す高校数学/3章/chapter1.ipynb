```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sympy import Symbol, solve

pd.set_option('display.max_rows', None)
```

```python
def plot(x, y, color='blue', axis='equal'):
    plt.plot(x, y, color=color)
    plt.axis(axis)
    plt.grid(color='0.8')
    plt.show()
```

```python
x = [1, 2, 3, 4, 5, 6, 7]
y = [64.3, 63.8, 63.6, 64.0, 63.5, 63.2, 63.1]

plt.plot(x, y)
plt.grid(color='0.8')
plt.show()
```

```python
def f(x):
    return x**3 - x

x = np.arange(-1, 1.01, 0.01)
y = np.array(f(x))

plot(x, y)
```

```python
def f1(x):
    return -2 * x + 7

def f2(x):
    return 1/2*x + 1/2

x = np.arange(-1, 6)
y1 = f1(x)
y2 = f2(x)

plt.plot(x, y1)
plt.plot(x, y2)
plt.axis('equal')

# plot(x, y1)
# plot(x, y2)
```

```python
x = Symbol('x')
y = Symbol('y')

ex1 = -3/2*x + 6 - y
ex2 = 1/2*x + 2 - y

print(solve((ex1, ex2)))
```

```python
th = np.arange(0, 360)
x = np.cos(np.radians(th))
y = np.sin(np.radians(th))

plot(x, y)
```
