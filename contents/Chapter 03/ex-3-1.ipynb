```python
#Âïè3.1
def q_3_1():
    vector = (-1, -2, 2)
    point = Points3D(vector)
    arrow = Arrow3D(vector)
    box = Box3D(*vector)
    draw3d(point, arrow, box)

q_3_1()
```

```python
#Âïè3.2
def q_3_2():
    pm1 = [1, -1]
    vertices = [(x, y, z) for x in pm1 for y in pm1 for z in pm1]
    edges = [] +\
        [((-1, y, z), (1, y, z)) for y in pm1 for z in pm1] +\
        [((x, -1, z), (x, 1, z)) for x in pm1 for z in pm1] +\
        [((x, y, -1), (x, y, 1)) for x in pm1 for y in pm1]
    
    
    points = [Points3D(v, color=blue) for v in vertices]
    segments = [Segment3D(*e, color=red) for e in edges]
    draw3d(*points, *segments)
    
    return

q_3_2()
```
