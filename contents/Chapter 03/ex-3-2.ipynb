```python
from draw3d import *
from math import sin, cos, pi
```

```python
def add(*vectors):
    by_coordinate = zip(*vectors)
    coordinate_sums = [sum(coords) for coords in by_coordinate]
    return tuple(coordinate_sums)

#問3.6
def scale(scalar, v):
    return tuple(scalar * n for n in v)

# 問3.3
def q_3_3():
    v1, v2 = (4, 0, 3), (-1, 0, 1)
    v3 = add(v1, v2)
    draw3d(
        Arrow3D(v1),
        Arrow3D(v3, v1),
        Arrow3D(v2, color=blue),
        Arrow3D(v3, v2, color=blue),
    )
    return

q_3_3()
```

```python
#問3.5
def q_3_5():
    size = 24
    vs = [(sin(pi*t/6), cos(pi*t/6), 1.0/3) for t in range(0, size)]
    
    running_sum = (0, 0, 0)
    arrows = []
    for v in vs:
        next_sum = add(running_sum, v)
        arrows.append(Arrow3D(next_sum, running_sum))
        running_sum = next_sum
    draw3d(*arrows)
    
    print('和 : ', running_sum)
    return

q_3_5()
```

```python
#問3.7
def q_3_7():
    u = (1, -1, -1)
    v = (0, 0, 2)
    return add(u, scale(1/2, add(v, scale(-1, u))))

q_3_7()
```

```python
#問3.9
def q_3_9(max_number=100):
    for x in range(1, max_number + 1):
        for y in range(1, x + 1):
            for z in range(1, y + 1):
                v = (x, y, z)
                distance = length(v)
                if distance.is_integer():
                    print('x: {x}, y: {y}, z: {z}, 長さ: {distance}'.format(x=x, y=y, z=z, distance=distance))
                
                
                
                
    return

q_3_9()
```
